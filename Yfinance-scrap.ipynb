````python
import pandas as pd
import yfinance as yf

# Mengambil data indeks NASDAQ dari Yahoo Finance
nasdaq = yf.Ticker("AAPL") # Apple Inc. Bitcoin(BTC-USD). NASDAQ ("^IXIC")

# Mengambil histori harga dengan periode maksimum
data = nasdaq.history(period="max")

# Reset index agar format lebih rapi
data.reset_index(inplace=True)

# Menampilkan 5 baris pertama
print(data.head())

# Simpan ke CSV
data.to_csv("./data/Apple_index.csv", index=False)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
sns.set_style("darkgrid")
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], label='Harga Penutupan AAPL')
````

````python
plt.figure(figsize=(12, 6))
plt.plot(df['Volume'], label='Volume Perdagangan AAPL', color='orange')
plt.title('Grafik Volume Perdagangan Saham Apple (AAPL) dari Waktu ke Waktu')
plt.xlabel('Tanggal')
plt.ylabel('Volume Perdagangan')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````
